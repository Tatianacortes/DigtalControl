1. La estadistica de este problema es la p:proporción de videos que se desarrollan en exteriores de la muetsra

*proporción: 37/128=0.29891 = 28.91%*
  
El parametro es la proporción real de todos los vídeos de YouTube que tienen alguna parte en exteriores, este valor es desconocido y debe estimarse con un intervalo de confianza.

2. - Estadística: (p̂) (proporción muestral)  
 - Parámetro: (p) (proporción de toda la población de vídeos de YouTube que se desarrollan en exteriores).

3. Para cada muestra bootstrap, se calcula una estaditica p̂, que representa la proporción de vídeos en exteriores en esa muestra particular.


4. El histograma de la distribución bootstrap de las proporciones muestrales p̂ debe estar centrado en la estadística muestral original

5. Para calcular el intervalo de confianza del 90 %, se toma el percentil 5 y el percentil 95 de la distribución bootstrap de p̂. Analizando el histograma se supone que los valores son:  
- Percentil 5: 0.22  
- Percentil 95: 0.36  

Entonces, el intervalo de confianza del 90 % es:  
\[
(0.22, 0.36)
\]


6. Se puede decir que se esta un 90 % seguro de que la proporción real de todos los vídeos de YouTube que tienen al menos una parte en exteriores está entre **0.22 y 0.36**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la muestra
n = 128  # Tamaño de la muestra
x = 37   # Número de videos en exteriores
p_hat = x / n  # Proporción muestral

# Simulación Bootstrap
num_bootstrap_samples = 1000
bootstrap_samples = np.random.binomial(n, p_hat, num_bootstrap_samples) / n

# Cálculo del intervalo de confianza del 90%
lower_bound = np.percentile(bootstrap_samples, 5)
upper_bound = np.percentile(bootstrap_samples, 95)

# Mostrar resultados
print(f"Proporción muestral (p̂): {p_hat:.4f}")
print(f"Intervalo de confianza del 90%: ({lower_bound:.4f}, {upper_bound:.4f})")

# Graficar la distribución bootstrap
plt.hist(bootstrap_samples, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(lower_bound, color='red', linestyle='dashed', label=f'5% Percentil ({lower_bound:.4f})')
plt.axvline(upper_bound, color='red', linestyle='dashed', label=f'95% Percentil ({upper_bound:.4f})')
plt.axvline(p_hat, color='blue', linestyle='solid', label=f'Proporción muestral ({p_hat:.4f})')
plt.xlabel('Proporción Bootstrap')
plt.ylabel('Frecuencia')
plt.title('Distribución Bootstrap de la Proporción Muestral')
plt.legend()
plt.show()
```

***7 PUNTO***

1.  El intervalo de confianza  del 95% para la proporción de adolescentes que han sufrido acoso cibernético es [0.54, 0.64], Como todo el intervalo está por encima de 0.50, significa que, con un 95% de confianza, más del 50% de los adolescentes han sufrido acoso.  

2. El valor 0.70 no está dentro del intervalo de confianza, lo que indica que el 70% no es una estimación plausible con un 95% de confianza.  

3. Un intervalo de confianza es del 90% es más reducido que el de uno de 95%, por lo que lo mas probable que el rango del intervlao este entre [0.55, 0.63], por lo que igualmente el 70% esteria a fuera del intervalo.



```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de una muestra basada en el intervalo [0.54, 0.64]
np.random.seed(42)
n = 1000  # Tamaño de la muestra simulada
p_medio = 0.59  # Punto medio del intervalo
simulated_data = np.random.binomial(n=1, p=p_medio, size=n)

# Bootstrap: generar 10,000 muestras y calcular proporciones
bootstrap_samples = 10000
bootstrap_proportions = np.random.choice(simulated_data, (bootstrap_samples, n), replace=True).mean(axis=1)

# Cálculo del intervalo de confianza del 95%
ci_95 = np.percentile(bootstrap_proportions, [2.5, 97.5])
print(f"Intervalo de confianza del 95%: {ci_95}")

# Verificación de las afirmaciones
print(f"¿El 50% está en el intervalo? {'Sí' if (0.50 >= ci_95[0] and 0.50 <= ci_95[1]) else 'No'}")
print(f"¿El 70% está en el intervalo? {'Sí' if (0.70 >= ci_95[0] and 0.70 <= ci_95[1]) else 'No'}")

# Intervalo de confianza del 90%
ci_90 = np.percentile(bootstrap_proportions, [5, 95])
print(f"Intervalo de confianza del 90%: {ci_90}")
print(f"¿El 70% está en el intervalo del 90%? {'Sí' if (0.70 >= ci_90[0] and 0.70 <= ci_90[1]) else 'No'}")

# Visualización del histograma de las proporciones bootstrap
plt.hist(bootstrap_proportions, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(ci_95[0], color='r', linestyle='dashed', label="Límite inferior 95%")
plt.axvline(ci_95[1], color='r', linestyle='dashed', label="Límite superior 95%")
plt.axvline(0.50, color='g', linestyle='solid', label="50% (afirmación del periódico)")
plt.axvline(0.70, color='b', linestyle='solid', label="70% (afirmación del investigador)")
plt.xlabel("Proporción bootstrap")
plt.ylabel("Frecuencia")
plt.title("Distribución Bootstrap de la proporción de acoso cibernético")
plt.legend()
plt.show()
```

1. 3 horas = 180 minutos

El intervalo de confianza es (128 min, 147 min), lo que significa que estamos un 95% seguros de que el tiempo promedio real está dentro de este rango, como 180 minutos está fuera del intervalo, **NO** se puede respaldar la afirmación del periódico.  
  
2. 2.2 horas = 132 minutos  
132 minutos está dentro del intervalo, lo que significa que este valor es plausible según los datos.  

3. Un intervalo de confianza del 99% es más amplio que uno del 95%, lo que significa que incluiría más valores extremos, Si 132 minutos ya estaba dentro del IC del 95%, entonces probablemente también estará dentro del intervalo de confianza del 99%**



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el intervalo de confianza del 95%
ic_95 = (128, 147)

# Definir los valores que queremos verificar
periodico = 180  # 3 horas en minutos
decano = 132  # 2.2 horas en minutos

# Verificaciones
periodico_respaldado = ic_95[0] <= periodico <= ic_95[1]
decano_respaldado = ic_95[0] <= decano <= ic_95[1]

print(f"¿El intervalo respalda la afirmación del periódico (180 min)? {'Sí' if periodico_respaldado else 'No'}")
print(f"¿El intervalo respalda la afirmación del decano (132 min)? {'Sí' if decano_respaldado else 'No'}")

# Simulación de un intervalo de confianza del 99% (más amplio)
ic_99 = (125, 150)  # Un IC del 99% es más amplio
decano_respaldado_99 = ic_99[0] <= decano <= ic_99[1]
print(f"¿El intervalo del 99% respalda la afirmación del decano (132 min)? {'Sí' if decano_respaldado_99 else 'No'}")

# Visualización
plt.figure(figsize=(8, 3))
plt.hlines(1, ic_95[0], ic_95[1], colors='r', label="IC 95% (128-147 min)", linewidth=3)
plt.hlines(2, ic_99[0], ic_99[1], colors='b', label="IC 99% (125-150 min)", linewidth=3)
plt.scatter([periodico, decano], [1, 1], color=['red', 'green'], label="Afirmaciones")
plt.text(periodico, 1, " Periódico (180 min)", verticalalignment='bottom', color='red')
plt.text(decano, 1, " Decano (132 min)", verticalalignment='bottom', color='green')

plt.ylim(0.5, 2.5)
plt.yticks([])
plt.xlabel("Tiempo de espera (min)")
plt.legend()
plt.title("Intervalos de confianza y afirmaciones")
plt.show()
```
